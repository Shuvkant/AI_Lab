```python
import matplotlib.pylab as plt
import numpy as np
```

```python
X=1.5
y=0.5
```

```python
W=np.random.randn(1)
print(W)
```

```python
epochs=20
alpha=0.1
errorList=[]
wList=[]

for epoch in range(epochs):
    # Forward Pass
    y_out=X*W

    #Compute the Error
    error=np.square((y-y_out))
    errorList.append(error)
    

    #compute the gradient of Error wrt w 
    gradient=-2*(y-y_out)*X
    

    #Backpropagate the error
    W=W-gradient*alpha
    wList.append(error)
print(W)
```

```python
plt.plot(range(epochs), errorList)
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.title("Epochs vs Error")
```
