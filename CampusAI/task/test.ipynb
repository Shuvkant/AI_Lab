```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a NumPy array
data = np.random.randn(100)

# Create a Pandas DataFrame
df = pd.DataFrame(data, columns=['Random Data'])

# Plot the data using Matplotlib
plt.figure(figsize=(10, 6))
plt.plot(df['Random Data'])
plt.title('Random Data Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```

```python
# Add a cumulative sum column
df['Cumulative Sum'] = df['Random Data'].cumsum()

# Add a rolling mean column with a window of 5
df['Rolling Mean'] = df['Random Data'].rolling(window=5).mean()

print(df.head())
```
